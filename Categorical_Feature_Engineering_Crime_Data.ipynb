**3. Dropping Redundant Columns:**

Upon reviewing the dataset, I have observed that the columns 'Crime Code' and 'Crime Code 1' appear to be similar. To determine the extent of their similarity, I am computing the percentage of similarity between these two columns. If the columns are found to contain identical data, one of them will be dropped to eliminate redundancy and streamline the dataset.

```python
#calculate the percentage of similarity between the columns Crime Code and Crime Code 1
def calculate_similarity(col1, col2):
    #ensure both columns are of the same length
    min_length = min(len(col1), len(col2))
    #compute the number of matching values
    matches = (col1[:min_length] == col2[:min_length]).sum()
    #calculate the similarity percentage
    similarity_percentage = (matches / min_length) * 100
    return similarity_percentage

similarity = calculate_similarity(data_cleaned['Crime Code'], data_cleaned['Crime Code 1'])
print(f"Similarity Percentage: {similarity:.2f}%")
```

The analysis revealed a 99.85% similarity between the columns 'Crime Code' and 'Crime Code 1'. The remaining 0.15% difference is attributed to missing values in 'Crime Code 1'. Based on this high similarity, the decision was made to retain 'Crime Code' and drop 'Crime Code 1' to reduce redundancy and maintain a streamlined dataset.

**4. Handling Missing Values in 'Modus Operandi Codes' column:**

To address missing values in the 'Modus Operandi Codes' column, I am investigating its relationship with 'Crime Code' and 'Weapon Used Code'. Given that 'Modus Operandi Codes' represent the methods or techniques used in committing a crime, while 'Crime Code' indicates the type of crime and 'Weapon Used Code' represents the weapon type, these columns may be interrelated. I will first analyze the most common combinations of these columns by grouping them. Based on this analysis, I will create a mapping to fill in the missing values in 'Modus Operandi Codes' using existing data.

```python
#check the relationship between Mocodes and Crime Code

#group by crime code and modus operandi codes
mo_crime_relation = data_df.groupby(['Crime Code', 'Modus Operandi Codes']).size().reset_index(name='count')

#sort by count to see the most common combinations
sorted_relation = mo_crime_relation.sort_values(by='count', ascending=False)

#display the top combinations
print(sorted_relation.head(10))
```

```python
#check the relationship between Mocodes and Weapon Used Code

#group by weapon used code and modus operandi codes
mo_weapon_relation = data_df.groupby(['Weapon Used Code', 'Modus Operandi Codes']).size().reset_index(name='count')

#sort by count to see the most common combinations
sorted_relation2 = mo_weapon_relation.sort_values(by='count', ascending=False)

#display the top combinations
print(sorted_relation2.head(10))
```

The analysis indicates that 'Crime Code' has more frequent and diverse combinations with the 'Modus Operandi Codes' column compared to 'Weapon Used Code'. Therefore, I will create a mapping from 'Crime Code' to 'Modus Operandi Codes' based on the existing data. This approach leverages the stronger correlation between these columns to effectively address the missing values in 'Modus Operandi Codes'.

```python
#create a mapping from Crime Code to Modus Operandi Codes based on existing data
crime_mo_mapping = mo_crime_relation.dropna(subset=['Modus Operandi Codes']).set_index('Crime Code')['Modus Operandi Codes'].to_dict()

#function to fill missing Modus Operandi Codes
def fill_mo_codes(row):
    if pd.isnull(row['Modus Operandi Codes']):
        return crime_mo_mapping.get(row['Crime Code'], row['Modus Operandi Codes'])
    return row['Modus Operandi Codes']

#apply the function to fill missing values in the 'Modus Operandi Codes' column
data_df['Modus Operandi Codes'] = data_df.apply(fill_mo_codes, axis=1)
```

```python
#check for remaining missing values in Modus Operandi Codes
missing_mocodes_count = data_df['Modus Operandi Codes'].isna().sum()
print(f"Remaining missing Modus Operandi Codes: {missing_mocodes_count}")
```

For the remaining missing values in the 'Modus Operandi Codes' column that were not addressed through internal mappings, I will utilize an external 'Modus Operandi Code' document from the LAPD. Given that the dataset pertains to Los Angeles, this external source will provide relevant codes. Since 'Crime Code' is closely related to 'Modus Operandi Code', I will use the crime code descriptions to match and fill in the missing 'Modus Operandi Codes', ensuring a more comprehensive and accurate completion of the dataset.

```python
#using an external file with modus operandi codes to fill the remaining null values

import re

#read the external file
external_file_path = './Downloads/mocodes.xlsx'
external_df = pd.read_excel(external_file_path)

#function to find the matching Modus Operandi Code
def find_matching_code(description, external_df):
    for _, row in external_df.iterrows():
        external_description = row['description']
        #check if any word in the external description matches the description in your dataset
        if any(word in description.split() for word in external_description.split()):
            return row['mocodes']
    return None

#function to fill missing Modus Operandi Codes
def fill_missing_mo_codes(row, external_df):
    if pd.isnull(row['Modus Operandi Codes']):
        return find_matching_code(row['Crime Code Description'], external_df)
    return row['Modus Operandi Codes']

#apply the function to fill missing values
data_df['Modus Operandi Codes'] = data_df.apply(fill_missing_mo_codes, axis=1, external_df=external_df)
```

```python
#check for remaining missing values in Modus Operandi Codes
missing_mocodes_count = data_df['Modus Operandi Codes'].isna().sum()
print(f"Remaining missing Modus Operandi Codes: {missing_mocodes_count}")
```

**10. Handling Missing Values in the 'Victim Sex' and 'Victim Descent' columns:**

To address the missing values in the 'Victim Sex' and 'Victim Descent' columns, I am using random values. I am choosing this approach because these columns may exhibit high variability and randomness can help avoid introducing bias. Random imputation ensures that the missing data does not disproportionately affect the distribution of these variables, while still allowing for a complete dataset for analysis.

```python
#distribution for 'Victim Sex'
sex_counts = data_df['Victim Sex'].value_counts(normalize=True)

#distribution for 'Victim Descent'
descent_counts = data_df['Victim Descent'].value_counts(normalize=True)

def fill_missing_values(df, column, distribution):
    missing_mask = df[column].isna()
    random_choices = np.random.choice(
        distribution.index, 
        size=missing_mask.sum(), 
        p=distribution.values
    )
    df.loc[missing_mask, column] = random_choices

#apply to 'Victim Sex'
fill_missing_values(data_df, 'Victim Sex', sex_counts)

#apply to 'Victim Descent'
fill_missing_values(data_df, 'Victim Descent', descent_counts)
```

```python
from datetime import time

#define a function to generate a random time as a string
def random_time():
    hour = np.random.randint(0, 24)
    minute = np.random.randint(0, 60)
    return time(hour=hour, minute=minute).strftime('%H:%M:%S')

#generate random times for each row in the DataFrame
sampled_data['Time Reported'] = [random_time() for _ in range(len(sampled_data))]
```

The 'Time Occurred' column currently displays time values as integers rather than in a time format. To address this, I will convert the existing integer values into a proper time format and create a new column for these formatted times. Subsequently, the original 'Time Occurred' column will be dropped. This process ensures that the time data is correctly formatted and suitable for accurate analysis.

```python
print(sampled_data['Time Occurred'].dtype)
```

```python
#define a function to convert integers to standardized time format
def convert_to_time_format(time_int):
    #ensure the integer is zero-padded to 4 digits
    time_str = f'{int(time_int):04}'
    #extract hours and minutes
    hours = time_str[:2]
    minutes = time_str[2:]
    #format as 'HH:MM:SS'
    return f'{hours}:{minutes}:00'

#ensure 'Time Occurred' is in integer format and handle non-integer values
sampled_data['Time Occurred'] = pd.to_numeric(sampled_data['Time Occurred'], errors='coerce')

#create a new column with the correct time format
sampled_data['New Time Occurred'] = sampled_data['Time Occurred'].apply(
    lambda x: convert_to_time_format(x) if pd.notna(x) else pd.NaT
)

#drop the original 'Time Occurred' column
sampled_data.drop(columns=['Time Occurred'], inplace=True)

#rename the new column to the original column name
sampled_data.rename(columns={'New Time Occurred': 'Time Occurred'}, inplace=True)
```

```python
#heatmap of top 15 crime types by area

#calculate the top 15 crime types
top_15_crimes = sampled_data['Crime Code Description'].value_counts().head(15).index

#filter data to include only the top 15 crimes
filtered_data = sampled_data[sampled_data['Crime Code Description'].isin(top_15_crimes)]

#calculate the count of each crime type in each area
area_crime_counts = filtered_data.pivot_table(index='Area Name', columns='Crime Code Description', aggfunc='size', fill_value=0)

#plotting
plt.figure(figsize=(16, 12))
sns.heatmap(area_crime_counts, cmap='Reds', annot=True, fmt='d')
plt.title('Heatmap of Top 15 Crime Types by Area')
plt.xlabel('Crime Type')
plt.ylabel('Area')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Analysis of Crime Distribution:**

The heatmap reveals that the area with the highest number of vehicle thefts is 77th Street, with Newton as the second most affected area. Additionally, Central is identified as the area with the highest incidence of burglary from vehicles. These insights are derived from the heatmap, which highlights crime concentration across different areas.

<font size="6">**Feature Engineering: Reducing Cardinality**</font>

**Reducing Cardinality:**

Cardinality is defined as the number of distinct values within a categorical feature, and lower cardinality is preferred because it simplifies the data, enhances model performance, and improves interpretability by reducing complexity. The 'Crime Code Description' and 'Weapon Description' columns exhibit high cardinality due to their numerous unique categories. To address this, I am employing the method of grouping rare categories into a single 'Other' category. This technique consolidates infrequent values, making the dataset more manageable and improving its analytical efficiency. Visualizations before and after this reduction are provided below to illustrate the impact of this approach on the data.

```python
#initial visualization #original high-cardinality 'Crime Code Description'
plt.figure(figsize=(10, 20))
sns.countplot(sampled_data['Crime Code Description'])
plt.title('Crime Code Description (Before Reduction)')
plt.xlabel('Count')
plt.ylabel('Crime Code Description')
plt.xticks(rotation=90)
plt.show()
```

```python
#group rare 'Crime Code Description' categories into 'Other'
top_crime_codes = sampled_data['Crime Code Description'].value_counts().nlargest(10).index
sampled_data['Crime Code Description'] = sampled_data['Crime Code Description'].apply(lambda x: x if x in top_crime_codes else 'OTHER')
```

```python
#reduced-cardinality 'Crime Code Description'
plt.figure(figsize=(12, 6))
sns.countplot(sampled_data['Crime Code Description'])
plt.title('Crime Code Description (After Reduction)')
plt.xlabel('Count')
plt.ylabel('Crime Code Description')
plt.xticks(rotation=90)
plt.show()
```

```python
#initial visualization #original high-cardinality 'Weapon Description'
plt.figure(figsize=(8, 16))
sns.countplot(sampled_data['Weapon Description'])
plt.title('Weapon Description (Before Reduction)')
plt.xlabel('Count')
plt.ylabel('Weapon Description')
plt.xticks(rotation=90)
plt.show()
```

```python
#group rare 'Weapon Description' categories into 'Other'
top_weapon_des = sampled_data['Weapon Description'].value_counts().nlargest(10).index
sampled_data['Weapon Description'] = sampled_data['Weapon Description'].apply(lambda x: x if x in top_weapon_des else 'OTHER')
```

```python
#reduced-cardinality 'Weapon Description'
plt.figure(figsize=(12, 6))
sns.countplot(sampled_data['Weapon Description'])
plt.title('Weapon Description (After Reduction)')
plt.xlabel('Count')
plt.ylabel('Weapon Description')
plt.xticks(rotation=90)
plt.show()
```

To prepare categorical variables for analysis, it is essential to choose an appropriate encoding method based on the nature of the data. For the 'Premises Description' column, I opted for Frequency Encoding, which assigns a numeric value based on the frequency of each category's occurrence in the dataset. This method is particularly useful when the frequency of categories may hold predictive significance, as it preserves the ordinal relationship between categories based on their frequency.

The code below implements Frequency Encoding by first calculating the frequency of each unique category in the 'Premises Description' column and then mapping these frequencies to a new column, 'Premises Description Encoded.' This approach helps retain the information inherent in the distribution of categories while converting the data into a format suitable for machine learning models. However, one limitation of Frequency Encoding is that it assumes the frequency of occurrence correlates with importance, which may not always be the case. Despite this, it can be an effective method when the frequency distribution is believed to be relevant to the analysis.

```python
#apply Frequency Encoding to 'Premises Description'
frequency_encoding = sampled_data['Premises Description'].value_counts()
sampled_data['Premises Description Encoded'] = sampled_data['Premises Description'].map(frequency_encoding)

#display the result
sampled_data[['Premises Description', 'Premises Description Encoded']].head()
```

To encode the categorical variables in my dataset, I chose Label Encoding for the 'Victim Sex' column. Label Encoding converts categorical text data into numerical values by assigning each unique category a different integer. This method is particularly suitable for columns where the categories have no specific order, but we still want to represent them numerically for machine learning models.

The primary advantage of Label Encoding is its simplicity and efficiency, especially when dealing with binary or low-cardinality categorical features like 'Victim Sex'. However, one limitation is that it can introduce ordinal relationships between categories where none exist, potentially leading to bias in models that assume numerical values imply a rank or order. In this context, where 'Victim Sex' has a limited and well-defined set of categories, Label Encoding is appropriate as it captures the distinctions without implying any unintended hierarchy.

```python
from sklearn.preprocessing import LabelEncoder

#initialize LabelEncoder
label_encoder = LabelEncoder()

#apply Label Encoding
sampled_data['Victim Sex'] = label_encoder.fit_transform(sampled_data['Victim Sex'])
```
